## Initialization

Let's create a cursor using the vertica_cursor function.

```python
from vertica_ml_python.utilities import vertica_cursor
cur = vertica_cursor("VerticaDSN")
```

During this study, we will work with a new dataset which is based on online retail. 

```python
from vertica_ml_python.utilities import drop_table
drop_table("retail", cur)
from vertica_ml_python.utilities import read_csv
retail = read_csv("retail.csv", cur)
```

We have access to many purchases of 4339 different customers based accross 38 different countries.

```python
print(retail)
```

<b>Question 1: </b>Build a dictionary of words and find the most relevant ones. What do you notice?

```python
from vertica_ml_python.learn.preprocessing import CountVectorizer
from vertica_ml_python.utilities import drop_text_index

drop_text_index("retail_voc", cur)

# We can build the model
model = CountVectorizer("retail_voc", cur)
model.fit("retail", ["Description"])

# We can export the dictionary
model.to_vdf().head(100)
```

<b>Question 2: </b>Build new features by extracting relevant words from the description.

```python
from vertica_ml_python import vDataframe
retail = vDataframe("retail", cur)
retail["Description"].apply("LOWER({})")
retail["Description"].add_copy("bag").str_extract('bag')["bag"].fillna(method = "0ifnull")
retail["Description"].add_copy("red").str_extract('red')["red"].fillna(method = "0ifnull")
retail["Description"].add_copy("heart").str_extract('heart')["heart"].fillna(method = "0ifnull")
retail["Description"].add_copy("retrospot").str_extract('retrospot')["retrospot"].fillna(method = "0ifnull")
retail["Description"].add_copy("design").str_extract('design')["design"].fillna(method = "0ifnull")
retail["Description"].add_copy("pink").str_extract('pink')["pink"].fillna(method = "0ifnull")
retail["Description"].add_copy("christmas").str_extract('christmas')["christmas"].fillna(method = "0ifnull")
retail["Description"].add_copy("box").str_extract('box')["box"].fillna(method = "0ifnull")
retail["Description"].add_copy("cake").str_extract('cake')["cake"].fillna(method = "0ifnull")
retail["Description"].add_copy("jumbo").str_extract('jumbo')["jumbo"].fillna(method = "0ifnull")
retail["Description"].add_copy("metal").str_extract('metal')["metal"].fillna(method = "0ifnull")
retail["Description"].add_copy("white").str_extract('white')["white"].fillna(method = "0ifnull")
retail["Description"].add_copy("blue").str_extract('blue')["blue"].fillna(method = "0ifnull")
retail["Description"].add_copy("lunch").str_extract('lunch')["lunch"].fillna(method = "0ifnull")
```
